```python
# Importing all libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10, 6
```

```python
#printing first five values of train CSV file
train =  pd.read_csv("Data/train_csv.csv")
train.head(2)
```

```python
#printing first five values of test CSV file
test = pd.read_csv("Data/test_csv.csv")
test.head(2)
```

```python
#Check the Datatypes of columns
train.info()
```

```python
import datetime
from datetime import datetime

#Parse object type to datetime type
train['time'] = pd.to_datetime(train['time'],format='%Y-%m-%d %H:%M:%S',infer_datetime_format=True)

#Set the Index
IndexedDataset = train.set_index(['time'])

#Drop the id column
IndexedDataset = IndexedDataset.drop(['id'], axis = 1) 

IndexedDataset.head(2)
```

```python
#Label the graph
plt.xlabel("Time in Hours, Minutes and Seconds on 19-03-2019")
plt.ylabel("Feature")
plt.grid(True)

#plot the graph
plt.plot(IndexedDataset)

# beautify the x-labels
plt.gcf().autofmt_xdate()
myFmt = mdates.DateFormatter('%H:%M:%S')
plt.gca().xaxis.set_major_formatter(myFmt)
```

According to the graph we can notice slight upward trend in the data

```python
#Determing Rolling Statistics
rolmean = IndexedDataset.rolling(window=24).mean()
rolstd = IndexedDataset.rolling(window=24).std()
print(rolmean,rolstd)
```

Here, we have taken a moving average rolling_window = 24.

It means an average to 24 timestamps would determine the prediction of seventh timestamp.

```python
#Plotting Rolling Statistics
original_data = plt.plot(IndexedDataset, color='blue', label='Original')
mean_data = plt.plot(rolmean, color='red', label='Rolling Mean')
std_data = plt.plot(rolstd, color='black', label='Rolling std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

Here your Mean and Standard deviation is not constant so you can conclude that your data is not stationary. Lets perform even Dickey-Fuller Test

```python
from statsmodels.tsa.stattools import adfuller
print('Results of Dickey-Fuller Test')
dftest = adfuller(IndexedDataset['feature'],autolag='AIC')
dfoutput = pd.Series(dftest[0:4],index=['Test Statistic','p-value','Number of lags used','Number of observations used'])
for key, value in dftest[4].items():
    dfoutput['Critical Value (%s)' %key] = value
print(dfoutput)
```

P-value should always be less but here we have 2.205 here we can say data is not stationary. So will estimate the trend

```python
#Estimate the trend
IndexedDataset_logScale = np.log(IndexedDataset)
plt.plot(IndexedDataset_logScale)
```

```python
movingaverage = IndexedDataset_logScale.rolling(window=24).mean()
movingstd = IndexedDataset_logScale.rolling(window=24).std()
plt.plot(IndexedDataset_logScale)
plt.plot(movingaverage,color='red')
```

```python
dls_minus_mv = IndexedDataset_logScale - movingaverage
dls_minus_mv.head(2)
```

```python
#Remove Nan Values
dls_minus_mv.dropna(inplace=True)
dls_minus_mv.head(2)
```

```python
from statsmodels.tsa.stattools import adfuller

def test_stationarity(timeseries):
    #Determing Rolling Statistics
    movingaverage = timeseries.rolling(window=24).mean()
    movingstd = timeseries.rolling(window=24).std()
    
    #Plot rolling statistics
    original_data = plt.plot(timeseries, color='blue', label='Original')
    mean_data = plt.plot(movingaverage, color='red', label='Rolling Mean')
    std_data = plt.plot(movingstd, color='black', label='Rolling std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller Test
    print('Results of Dickey-Fuller Test')
    dftest = adfuller(timeseries['feature'],autolag='AIC')
    dfoutput = pd.Series(dftest[0:4],index=['Test Statistic','p-value','Number of lags used','Number of observations used'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)' %key] = value
    print(dfoutput)
    
```

```python
test_stationarity(dls_minus_mv)
```

```python
exponentialDecayWeightedAverage = IndexedDataset_logScale.ewm(halflife=12,min_periods=0,adjust=True).mean()
plt.plot(IndexedDataset_logScale)
plt.plot(exponentialDecayWeightedAverage, color='red')
```

```python
dls_minus_meda = IndexedDataset_logScale - exponentialDecayWeightedAverage
test_stationarity(dls_minus_meda)
```

```python
datasetLogDiffShiffting = IndexedDataset_logScale - IndexedDataset_logScale.shift()
plt.plot(datasetLogDiffShiffting)
```

```python
datasetLogDiffShiffting.dropna(inplace=True)
test_stationarity(datasetLogDiffShiffting)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(IndexedDataset_logScale, period=24)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(IndexedDataset_logScale, label='Original')
plt.legend(loc = 'best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc = 'best')
plt.subplot(413)
plt.plot(seasonal, label='Seasonality')
plt.legend(loc = 'best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc = 'best')
plt.tight_layout()
```

```python
from statsmodels.tsa.stattools import acf, pacf 

lag_acf = acf(datasetLogDiffShiffting, nlags=25) 
lag_pacf = pacf(datasetLogDiffShiffting, nlags=25, method='ols')

#Plot ACF
plt.subplot(122)
plt.plot(lag_acf) 
plt.axhline(y=0,linestyle='--',color='gray') 
plt.axhline(y=-1.96/np.sqrt(len(datasetLogDiffShiffting)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(datasetLogDiffShiffting)),linestyle='--',color='gray') 
plt.title('Autocorrelation Function') 
plt.show()

#Plot PACF
plt.subplot(122)
plt.plot(lag_pacf) 
plt.axhline(y=0,linestyle='--',color='gray') 
plt.axhline(y=-1.96/np.sqrt(len(datasetLogDiffShiffting)),linestyle='--',color='gray') 
plt.axhline(y=1.96/np.sqrt(len(datasetLogDiffShiffting)),linestyle='--',color='gray') 
plt.title('Partial Autocorrelation Function')
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA

#AR Model
model = ARIMA(IndexedDataset_logScale, order=(2, 1, 0))  
# here the q value is zero  

results_AR = model.fit(disp=-1)  
plt.plot(datasetLogDiffShiffting, label='original') 
plt.plot(results_AR.fittedvalues, color='red', label='predictions') 
plt.title('RSS: %.4f'%sum((results_AR.fittedvalues-datasetLogDiffShiffting["feature"])**2)) 
print('Plotting AR model')

plt.show()
```

```python
#MA Model
model = ARIMA(IndexedDataset_logScale, order=(0, 1, 1))  
# here the p value is zero 

results_MA = model.fit(disp=-1)  
plt.plot(datasetLogDiffShiffting) 
plt.plot(results_MA.fittedvalues, color='red') 
plt.title('RSS: %.4f'%sum((results_MA.fittedvalues-datasetLogDiffShiffting["feature"])**2)) 
print('Plotting MA model')

plt.show()
```

```python
model = ARIMA(IndexedDataset_logScale, order=(2, 1, 2))  

results_ARIMA = model.fit(disp=-1)  
plt.plot(datasetLogDiffShiffting) 
plt.plot(results_ARIMA.fittedvalues, color='red') 
plt.title('RSS: %.4f'%sum((results_ARIMA.fittedvalues-datasetLogDiffShiffting["feature"])**2)) 

plt.show()
```
